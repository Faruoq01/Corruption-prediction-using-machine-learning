```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

# Read data from excel file

```python
data = pd.read_csv('Corruption.csv')
data.head(10)
```

# Extract input and output data

```python
A = data['A']
B = data['B']
C = data['C']
D = data['D']
E = data['E']
F = data['F']
G = data['G']
H = data['H']
I = data['I']
J = data['J']
```

# Getting inputs and outputs data for machine learning

```python
inputs = np.column_stack((A, B, C, D, E, F, G, H, I, J))
```

```python
targets = np.column_stack((data['outputs'],))
```

# Initializing data for machine learning

```python
weights = np.column_stack((np.random.uniform(-0.2, 0.2, size=(1, 10))))
Bias = np.random.uniform(-0.2, 0.2, size=1)
Learning_rate = 0.009
observations = 117
```

```python
plt.scatter(A, targets)
```

# Training the data

```python
for i in range(500):
    outputData = np.dot(inputs, weights) + Bias
    delta = outputData - targets
    Loss = np.sum(delta**2/observations)
    print(Loss)
    delta_scaled = 2*delta/observations
    
    #update rules
    Weights = weights - Learning_rate*np.dot(inputs.T, delta_scaled)
    Bias = Bias - Learning_rate*np.sum(delta_scaled)
```
